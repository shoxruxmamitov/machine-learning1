skill

```python
from sklearn.base import BaseEstimator, TransformerMixin

roms_x, bedroom_x, population_x, households_x = 3, 4, 5, 6

class CombinedAttributesAdder(BaseEstimator, TransformerMixin):
    def __init__(self, add_bedrooms_per_room=True):
        self.add_bedrooms_per_room = add_bedrooms_per_room
    def fit(self, X, y=None):
        return self
    def transform(self, X):
      roms_per_household = X[:, roms_x] / X[:, households_x]
      population_per_household = X[:, population_x] / X[:, households_x]
      if self.add_bedrooms_per_room:
        bedrooms_per_room = X[:, bedroom_x] / X[:, roms_x]
        return np.c_[X, roms_per_household, population_per_household, bedrooms_per_room]
      else:
        return np.c_[X, roms_per_household, population_per_household]
```
